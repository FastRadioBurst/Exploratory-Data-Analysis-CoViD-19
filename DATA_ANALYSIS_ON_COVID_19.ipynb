The severity of COVID has increased resulting in more deaths. We tried to compare the severity of COVID with other dieases such as Pnuemonia and Influenza.

The following subplot shows the comparison of death count along with the age distribution. This plot also potrays that 65 and above are remarkably effected by Pneumonia and Influenza.



```
#Libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#Grid and figure size
sns.set(style="whitegrid")
f, axes = plt.subplots(2, 2, figsize=(16, 10), sharex=True)
#subplots
plot1 = sns.barplot(x="Age_group", y="COVID Deaths", data=df1, palette = 'YlOrRd', edgecolor ='black',ax = axes[0,0]).set_title("COVID Deaths")
plot2 = sns.barplot(x="Age_group", y="Pneumonia Deaths", data=df1, palette = 'YlOrRd', edgecolor ='black', ax=axes[0,1]).set_title("Pneumonia Deaths")
plot3 = sns.barplot(x="Age_group", y="Influenza Deaths", data = df1,palette = 'YlOrRd',edgecolor ='black', ax= axes[1,0]).set_title("Influenza Deaths")
plot4 = sns.barplot(x="Age_group", y="Pneumonia_ COVID", data = df1,palette = 'YlOrRd',edgecolor='black', ax= axes[1,1]).set_title("Pneumonia and COVID Deaths")
#Label Rotation
for ax in axes.flatten():
    for label in ax.get_xticklabels():
        label.set_rotation(90)

plt.show()
```
